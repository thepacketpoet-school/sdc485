```python
# Import necessary libraries
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report
import xgboost as xgb
from datetime import datetime
from IPython.display import display, Markdown
todays_date = datetime.now().strftime("%Y-%m-%d")
student_id = "halarc1407"
# Load the Breast Cancer dataset
data = pd.read_csv('breastCancerDataset.csv')
# Display the first few rows
display(Markdown("### Step 1.2: Load the Dataset"))
display(Markdown(f"Date: {todays_date}"))
display(Markdown(f"Student ID: {student_id}"))
display(data.head())
```

```python
display(Markdown("### Step 1.3: Initial Exploration of the Dataset"))
display(Markdown(f"Date: {todays_date}"))
display(Markdown(f"Student ID: {student_id}"))
# Concise summary of the DataFrame
print(data.info())
# Descriptive statistics
display(data.describe())
```

```python
display(Markdown("### Step 1.4: Preprocess the Data"))
display(Markdown(f"Date: {todays_date}"))
display(Markdown(f"Student ID: {student_id}"))
# Check and handle missing values with the median
print(f"Missing values before:\n{data.isnull().sum().sum()}")
for column in data.select_dtypes(include=[np.number]).columns:
    data[column] = data[column].fillna(data[column].median())
print(f"Missing values after: {data.isnull().sum().sum()}")

# Identify and remove duplicate rows
print(f"\nDuplicate rows before: {data.duplicated().sum()}")
data = data.drop_duplicates()
print(f"Duplicate rows after: {data.duplicated().sum()}")

# Standardize the concavity1 and concavity3 features
scaler = StandardScaler()
data[['concavity1', 'concavity3']] = scaler.fit_transform(data[['concavity1', 'concavity3']])
print("\nconcavity1 and concavity3 standardized.")

# Encode the Diagnosis target (M=1, B=0)
data['Diagnosis'] = data['Diagnosis'].map({'M': 1, 'B': 0})

# Split features and target, drop ID
X = data.drop(['Diagnosis', 'ID'], axis=1)
y = data['Diagnosis']

# Split into training (70%) and testing (30%) sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
print(f"\nTraining set size: {X_train.shape}")
print(f"Testing set size: {X_test.shape}")
```

```python
display(Markdown("### Step 2.1: Implement SVM"))
display(Markdown(f"Date: {todays_date}"))
display(Markdown(f"Student ID: {student_id}"))
# Fit the SVM model
svm = SVC()
svm.fit(X_train, y_train)
y_pred_svm = svm.predict(X_test)
# Evaluate the SVM model
print(f"SVM Accuracy Score: {accuracy_score(y_test, y_pred_svm):.4f}")
print("\nClassification Report:")
print(classification_report(y_test, y_pred_svm))
```

```python
display(Markdown("### Step 2.2: GridSearchCV for SVM Hyperparameter Tuning"))
display(Markdown(f"Date: {todays_date}"))
display(Markdown(f"Student ID: {student_id}"))
# Define the parameter grid
param_grid_svm = {
    'C': [0.1, 1, 10, 100],
    'kernel': ['linear', 'rbf', 'poly']
}
# Use GridSearchCV
grid_search_svm = GridSearchCV(SVC(), param_grid_svm, cv=5)
grid_search_svm.fit(X_train, y_train)
print(f"Best Parameters: {grid_search_svm.best_params_}")
print(f"Best Cross-Validation Score: {grid_search_svm.best_score_:.4f}")
```

```python
display(Markdown("### Step 2.3: Evaluate the Tuned SVM Model"))
display(Markdown(f"Date: {todays_date}"))
display(Markdown(f"Student ID: {student_id}"))
# Predict with the best SVM model
best_svm = grid_search_svm.best_estimator_
y_pred_svm_tuned = best_svm.predict(X_test)
# Evaluate
print(f"Tuned SVM Accuracy Score: {accuracy_score(y_test, y_pred_svm_tuned):.4f}")
print("\nClassification Report:")
print(classification_report(y_test, y_pred_svm_tuned))
```

```python
display(Markdown("### Step 3.1: Implement XGBoost"))
display(Markdown(f"Date: {todays_date}"))
display(Markdown(f"Student ID: {student_id}"))
# Fit the XGBoost model
xgb_model = xgb.XGBClassifier(eval_metric='logloss')
xgb_model.fit(X_train, y_train)
y_pred_xgb = xgb_model.predict(X_test)
# Evaluate the XGBoost model
print(f"XGBoost Accuracy Score: {accuracy_score(y_test, y_pred_xgb):.4f}")
print("\nClassification Report:")
print(classification_report(y_test, y_pred_xgb))
```

```python
display(Markdown("### Step 3.2: GridSearchCV for XGBoost Hyperparameter Tuning"))
display(Markdown(f"Date: {todays_date}"))
display(Markdown(f"Student ID: {student_id}"))
# Define the parameter grid
param_grid_xgb = {
    'n_estimators': [50, 100],
    'learning_rate': [0.1, 0.2],
    'max_depth': [3, 4]
}
# Use GridSearchCV
grid_search_xgb = GridSearchCV(xgb.XGBClassifier(eval_metric='logloss'), param_grid_xgb, cv=5)
grid_search_xgb.fit(X_train, y_train)
print(f"Best Parameters: {grid_search_xgb.best_params_}")
print(f"Best Cross-Validation Score: {grid_search_xgb.best_score_:.4f}")
```

```python
display(Markdown("### Step 3.3: Evaluate the Tuned XGBoost Model"))
display(Markdown(f"Date: {todays_date}"))
display(Markdown(f"Student ID: {student_id}"))
# Predict with the best XGBoost model
best_xgb = grid_search_xgb.best_estimator_
y_pred_xgb_tuned = best_xgb.predict(X_test)
# Evaluate
print(f"Tuned XGBoost Accuracy Score: {accuracy_score(y_test, y_pred_xgb_tuned):.4f}")
print("\nClassification Report:")
print(classification_report(y_test, y_pred_xgb_tuned))
```
