```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
from IPython.display import display, Markdown

todays_date = datetime.now().strftime('%Y-%m-%d')
student_id = 'halarc1407'

display(Markdown('### Set Up the Simulation'))
display(Markdown(f'Date: {todays_date}'))
display(Markdown(f'Student ID: {student_id}'))

# Define simulation parameters
growth_rate = 0.1
initial_population = 100
time_steps = 50

print(f'Growth Rate:        {growth_rate}')
print(f'Initial Population: {initial_population}')
print(f'Time Steps:         {time_steps}')
```

```python
display(Markdown('### Implement and Run the Simulation'))
display(Markdown(f'Date: {todays_date}'))
display(Markdown(f'Student ID: {student_id}'))

def simulate_population_growth(initial_population, growth_rate, time_steps):
    """Simulate population growth using the given parameters."""
    population = initial_population
    results = [population]
    for _ in range(time_steps):
        population = population + (population * growth_rate)
        results.append(population)
    return results

population_data = simulate_population_growth(initial_population, growth_rate, time_steps)
time = list(range(time_steps + 1))

print(f'Starting population: {population_data[0]:.2f}')
print(f'Final population:    {population_data[-1]:.2f}')
```

```python
display(Markdown('### Analyze the Simulation Results'))
display(Markdown(f'Date: {todays_date}'))
display(Markdown(f'Student ID: {student_id}'))

pop_series = pd.Series(population_data)
print('Population Growth Summary Statistics:')
print(f'  Min:    {pop_series.min():.2f}')
print(f'  Max:    {pop_series.max():.2f}')
print(f'  Mean:   {pop_series.mean():.2f}')
print(f'  Median: {pop_series.median():.2f}')
print(f'  Std:    {pop_series.std():.2f}')
```

```python
display(Markdown('### Visualize the Results'))
display(Markdown(f'Date: {todays_date}'))
display(Markdown(f'Student ID: {student_id}'))

plt.figure(figsize=(12, 6))
plt.plot(time, population_data, marker='o', markersize=3, color='steelblue', label=f'Growth Rate = {growth_rate}')
plt.title('Population Growth Over Time')
plt.xlabel('Time Step')
plt.ylabel('Population')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
display(Markdown('### Apply Sensitivity Analysis'))
display(Markdown(f'Date: {todays_date}'))
display(Markdown(f'Student ID: {student_id}'))

# Vary growth rate to observe impact
sensitivity_rates = [0.05, 0.10, 0.15, 0.20, 0.25]

plt.figure(figsize=(12, 6))
for rate in sensitivity_rates:
    pop = simulate_population_growth(initial_population, rate, time_steps)
    linestyle = '-' if rate == 0.10 else '--'
    label = f'Rate = {rate}' + (' (original)' if rate == 0.10 else '')
    plt.plot(time, pop, linestyle=linestyle, label=label)

plt.title('Sensitivity Analysis: Impact of Growth Rate on Population')
plt.xlabel('Time Step')
plt.ylabel('Population')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

print('Final population by growth rate:')
for rate in sensitivity_rates:
    pop = simulate_population_growth(initial_population, rate, time_steps)
    print(f'  Rate {rate}: {pop[-1]:,.2f}')
```
