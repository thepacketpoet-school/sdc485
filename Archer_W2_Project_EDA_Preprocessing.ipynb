```python
# Import necessary libraries
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from datetime import datetime
from IPython.display import display, Markdown
todays_date = datetime.now().strftime("%Y-%m-%d")
student_id = "halarc1407"
# Load the dataset
data = pd.read_csv('Credit_card_data.csv')
display(Markdown("### Step 1: Load the Dataset"))
display(Markdown(f"Date: {todays_date}"))
display(Markdown(f"Student ID: {student_id}"))
display(data.head())
print(f"\nDataset shape: {data.shape}")
```

```python
display(Markdown("### Step 2: Initial Dataset Information"))
display(Markdown(f"Date: {todays_date}"))
display(Markdown(f"Student ID: {student_id}"))
# Concise summary of the DataFrame
print(data.info())
print("\nMissing values per column:")
print(data.isnull().sum())
```

## Data Preprocessing

```python
display(Markdown("### Step 9: Handle Missing Values"))
display(Markdown(f"Date: {todays_date}"))
display(Markdown(f"Student ID: {student_id}"))
# Make a copy for preprocessing so we preserve the original
data_processed = data.copy()

# Fill numerical missing values with median (more robust to outliers than mean)
# reports has high skew, active has outliers, so median is the safer choice
for column in ['reports', 'dependents', 'active']:
    median_val = data_processed[column].median()
    data_processed[column] = data_processed[column].fillna(median_val)
    print(f"Filled {column} missing values with median: {median_val}")

# selfemp is categorical, fill with mode (most common value)
mode_val = data_processed['selfemp'].mode()[0]
data_processed['selfemp'] = data_processed['selfemp'].fillna(mode_val)
print(f"Filled selfemp missing values with mode: {mode_val}")

print(f"\nRemaining missing values:\n{data_processed.isnull().sum()}")
```

```python
display(Markdown("### Step 10: One-Hot Encoding of Categorical Features"))
display(Markdown(f"Date: {todays_date}"))
display(Markdown(f"Student ID: {student_id}"))
# Encode the binary categorical variables: card, owner, selfemp
# Using drop_first=True to avoid the dummy variable trap
data_processed = pd.get_dummies(data_processed, columns=['card', 'owner', 'selfemp'], 
                                 drop_first=True, dtype=int)
print("Columns after one-hot encoding:")
print(data_processed.columns.tolist())
display(data_processed.head())
```

```python
display(Markdown("### Step 11: Feature Scaling (Standardization)"))
display(Markdown(f"Date: {todays_date}"))
display(Markdown(f"Student ID: {student_id}"))
# Standardize numerical features for neural network compatibility
# Neural networks and distance-based methods (like K-means clustering) need scaled features
numerical_cols_to_scale = ['reports', 'age', 'income', 'share', 'expenditure', 
                            'dependents', 'months', 'majorcards', 'active']
scaler = StandardScaler()
data_processed[numerical_cols_to_scale] = scaler.fit_transform(data_processed[numerical_cols_to_scale])
print("Numerical features after standardization (mean ~0, std ~1):")
display(data_processed[numerical_cols_to_scale].describe().round(3))
```

```python
display(Markdown("### Step 12: Bin the Age Feature"))
display(Markdown(f"Date: {todays_date}"))
display(Markdown(f"Student ID: {student_id}"))
# Create age bins from the original (unscaled) data for interpretability
# This is useful for the clustering and customer segmentation analytical task
age_bins = pd.cut(data['age'], bins=[0, 25, 35, 50, 100], 
                  labels=['Under 25', '25-35', '35-50', 'Over 50'])
print("Age bin distribution:")
print(age_bins.value_counts())
# One-hot encode the age bins and add to processed data
age_bin_dummies = pd.get_dummies(age_bins, prefix='age_group', dtype=int)
data_processed = pd.concat([data_processed, age_bin_dummies], axis=1)
print("\nFinal preprocessed dataset shape:", data_processed.shape)
display(data_processed.head())
```

```python
display(Markdown("### Step 13: Save Preprocessed Dataset"))
display(Markdown(f"Date: {todays_date}"))
display(Markdown(f"Student ID: {student_id}"))
# Save the preprocessed dataset for use in the next phase (neural network modeling)
data_processed.to_csv('Credit_card_data_processed.csv', index=False)
print("Preprocessed dataset saved as 'Credit_card_data_processed.csv'")
print(f"Final shape: {data_processed.shape}")
```
